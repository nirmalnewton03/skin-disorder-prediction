- This project analyzes skin disorder data to identify key patterns and trends in skin disorder cases. A predictive model will be developed using various machine learning classifiers to classify patients accurately by type of skin disorder.

# DOMAIN ANALYSIS AND ATTRIBUTES

This project falls under **"Healthcare"** domain to classify patients accurately by type of skin disorder based on the key attributes such as,

## Clinical Attributes:

-**Erythema**: Redness of the skin, often due to inflammation. It is a common symptom in many skin diseases, including psoriasis and eczema.

-**Scaling**: The shedding of the outer layer of the skin, often seen in conditions like psoriasis and ichthyosis.

-**Definite Borders**: Clearly defined edges of skin lesions, which can help differentiate between diseases like psoriasis (well-defined borders) and eczema (less defined).

-**Itching**: Pruritus, a common symptom in conditions such as eczema and lichen planus.

-**Koebner Phenomenon**: The appearance of skin lesions at the site of trauma, seen in conditions like psoriasis and lichen planus.

-**Polygonal Papules**: Small, raised bumps with flat tops, often seen in lichen planus.

-**Follicular Papules**: Bumps centered around hair follicles, commonly associated with conditions like follicular eczema or keratosis pilaris.

-**Oral Mucosal Involvement**: Involvement of the mucous membranes inside the mouth, seen in conditions like lichen planus.

-**Knee and Elbow Involvement**: Common sites for psoriasis lesions.

-**Scalp Involvement**: Involvement of the scalp, common in conditions like psoriasis and seborrheic dermatitis.

-**Family History**: Presence of similar skin conditions in family members, relevant for conditions like psoriasis.

## Histopathological Attributes:

-**Melanin Incontinence**: Disruption of melanin pigment into the dermis, often seen in conditions causing inflammation or trauma.

-**Eosinophils in the Infiltrate**: Presence of eosinophils, a type of white blood cell, indicating allergic reactions or parasitic infections.

-**PNL Infiltrate**: Presence of polymorphonuclear leukocytes (PMNs), indicating acute inflammation.

-**Fibrosis of the Papillary Dermis**: Scarring in the upper dermis, often seen in chronic conditions like lichen planus.

-**Exocytosis**: Movement of cells into the epidermis from the dermis, seen in conditions like eczema.

-**Acanthosis**: Thickening of the epidermis, common in conditions like psoriasis.

-**Hyperkeratosis**: Thickening of the stratum corneum, often seen in conditions like ichthyosis.

-**Parakeratosis**: Presence of nuclei in the stratum corneum, indicating abnormal maturation of skin cells.

-**Clubbing of the Rete Ridges**: Distortion of the rete ridges, seen in conditions like psoriasis.

-**Elongation of the Rete Ridges**: Lengthening of the rete ridges, often seen in conditions like psoriasis.

-**Thinning of the Suprapapillary Epidermis**: Reduction in thickness of the epidermis above the dermal papillae, seen in conditions like psoriasis.

-**Spongiform Pustule**: A type of pustule seen in conditions like pustular psoriasis.

-**Munro Microabscess**: Collections of neutrophils in the stratum corneum, characteristic of psoriasis.

-**Focal Hypergranulosis**: Increased thickness of the granular layer in certain areas, seen in conditions like psoriasis.

-**Disappearance of the Granular Layer**: Loss of the granular layer, often seen in conditions like psoriasis.

-**Vacuolisation and Damage of Basal Layer**: Damage to the basal layer of the epidermis, seen in conditions like lichen planus.

-**Spongiosis**: Edema within the epidermis, often seen in conditions like eczema.

-**Saw-Tooth Appearance of Rete Ridges**: Distortion of the rete ridges, characteristic of psoriasis.

-**Follicular Horn Plug**: A plug of keratin within a hair follicle, seen in conditions like keratosis pilaris.

-**Perifollicular Parakeratosis**: Presence of parakeratotic cells around hair follicles, seen in conditions like seborrheic dermatitis.

-**Inflammatory Mononuclear Infiltrate**: Presence of mononuclear cells (like lymphocytes and macrophages) indicating chronic inflammation.

-**Band-Like Infiltrate**: A dense band of inflammatory cells at the dermo-epidermal junction, characteristic of lichen planus.

-**Age**: A linear attribute indicating the patient's age, which can influence the likelihood of certain skin conditions.

## Target attribute:

-**Class**: Patients with various types of skin disorders(1:psoriasis, 2:seboreic dermatitis, 3:lichen planus, 4:pityriasis rosea, 5:cronic dermatitis and 6:pityriasis rubra pilaris).

# 7. DATA SPLITTING

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

### SMOTE 

```python
from imblearn.over_sampling import SMOTE
smote= SMOTE(random_state=42)
x_train_smote,y_train_smote=smote.fit_resample(x_train,y_train)
```

```python
from collections import Counter

print("Class distribution before SMOTE:", Counter(y_train)) 
print("Class distribution after SMOTE:", Counter(y_train_smote))  
```

### FEATURE EXTRACTION

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=20)
pca_x_train=pca.fit_transform(x_train_smote)
pca_x_test=pca.transform(x_test)
print(pca_x_train.shape)
print(pca_x_test.shape)
```

```python
original_features = x_train_smote.columns
pca_loadings = pd.DataFrame(np.abs(pca.components_), columns=original_features)
important_features = pca_loadings.idxmax(axis=1)
print("Original features that contribute the most to each principal component:")
print(important_features)
```
